```
import os
import numpy as np
import pandas as pd
import json
import ast
from pandas.io.json import json_normalize

path = os.path.join(os.getcwd(),'drive/My Drive/task1.csv')
```

```
df = pd.read_csv(path)
df
```

```
df = pd.concat([df['User ID'],df['Response']],axis=1)
df
```

```
print(df['Response'][2])
```

```
df['Response'] = df['Response'].apply(lambda x : ast.literal_eval(str(x)))
```

```
df
```

```
df2 = json_normalize(df['Response'])
df2
```

```
df3 = pd.concat([df['User ID'],df2['data']],axis=1)
df3
```

```
df3 = df3[df3['data'].notnull()]
df3
```

```
df3.index = pd.RangeIndex(start = 0 , stop=319,step=1)
df3
```

```
bank_names = []
for i in range(0,318):
  bank_names.append(df3['data'][i][0]['bankName'])
bank_names
```

```
df3['data'][0][0]
```

```
account_nos = []
for i in range(0,318):
  account_nos.append(df3['data'][i][0]['accountNumber'])
account_nos
```

```
ac_dict = pd.DataFrame({'account_nos' : account_nos})
bn_dict = pd.DataFrame({'back_names' : bank_names})
```

```
df_final = pd.concat([df3['User ID'],bn_dict,ac_dict],axis=1)
df_final
```

```
df_final.to_csv('final.csv',index=False)
```
