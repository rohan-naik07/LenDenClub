```
import numpy as np
import pandas as pd
import ast
import json
import os
from pandas.io.json import json_normalize  

path = '/content/drive/My Drive/task1.csv'
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
dataframe = pd.read_csv(path, usecols = ['User ID','Response'])
```

```
dataframe['Response'] = dataframe['Response'].apply(lambda data_entry : ast.literal_eval(str(data_entry)))   
```

```
print(type(dataframe['Response'][0]))
dataframe
```

```
dataframe2 = json_normalize(dataframe['Response'])
dataframe2
```

```
dataframe2 = pd.concat([dataframe['User ID'],dataframe2['data']],axis=1)
dataframe2
```

```
dataframe2 = dataframe2[dataframe2['data'].notnull()]
```

```
dataframe2.index = pd.RangeIndex(start = 0,stop=dataframe2.shape[0],step=1)
```

```
dataframe2
```

```
dataframe2['data'][0]
```

```
sample_date = dataframe2['data'][2][0]['salary'][0]['transactions'][0]['transactionDate']
```

```
import datetime 
your_dt = datetime.datetime.fromtimestamp(int(sample_date)/1000).strftime("%Y-%m-%d %H:%M:%S")
your_dt
```

```
bank_names = []
account_nos =[]
account_name =[]
salary = []
salary_data = []

for data_index in range(dataframe2.shape[0]):
  if dataframe2['data'][data_index]:
    bank_names.append(dataframe2['data'][data_index][0]['bankName'])
    account_nos.append(dataframe2['data'][data_index][0]['accountNumber'])
    account_name.append(dataframe2['data'][data_index][0]['accountName'])
    salary_data.append( dataframe2['data'][data_index][0]['salary'])
    if dataframe2['data'][data_index][0]['salary']:
      salary.append(dataframe2['data'][data_index][0]['salary'][-1]['totalSalary'])
    else:
      salary.append(-1)
```

```
salary_data[0]
```

```
salary_dates = []
for salary in salary_data:
  for data in salary:
    sample_date = data['transactions'][0]['transactionDate']
    salary_dates.append(datetime.datetime.fromtimestamp(int(sample_date)/1000).strftime("%Y-%m-%d %H:%M:%S"))

len(salary_dates)
```
